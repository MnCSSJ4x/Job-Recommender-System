```python
infos = []
for page in range(2, 41):
        # get the jobs list items to scroll through:
        jobs = driver.find_elements(By.CLASS_NAME, "occludable-update")
        print(len(jobs) - 1)
        for i in range(len(jobs) - 1):
            job = jobs[i]
            # scroll_to(job)
            driver.execute_script("arguments[0].scrollIntoView();", job)
            job.click()
            # [position, company, location, details] = self.get_position_data(job)
            # infos.append(get_position_data(job))
            l = job.text.split('\n')
            print(l)
            [position, company, location] = job.text.split('\n')[:3]
            details = driver.find_element("id", "job-details").text
            infos.append([position, company, location, details])
            # append rows to an empty DataFrame
            df = df.append({'Position' : position, 'Company' : company, 'Location' : location, 'Job Description' : details}, ignore_index = True)

            print("\n" + position + " " + company + " " + location + "\n")
            # do something with the data...
        # go to next page:
        driver.find_element('xpath', f"//button[@aria-label='Page {page}']").click()
```
